```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x1 = np.random.randint(1,50,30)
x1 = np.sort(x1)#It will sort the array is asceending 
x2= np.random.randint(1000,70000,30)
plt.plot(x1,x2)
```

```python
#2. After Nomalization 
x1min = min(x1)
x1max = max(x1)
x2min = min(x2)
x2max = max(x2)
x1norm = (x1 - x1min)/(x1max - x1min)
x2norm = (x2 - x2min)/(x2max - x2min)
plt.plot(x1norm,x2norm)
```

```python
#DATASET - SOCIAL NETWORKS ADS
#DATASET https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Social_Network_Ads.csv
```

```python
#1. Take the data create data frame
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Social_Network_Ads.csv')
df
```

```python
#purchase : 0 and 1
#0 not purchased
#1 purchased
#output =  purchased
#input = age & salary
```

```python
#i just want to know how many people how many people have purchased and how many have not purchased
df['Purchased'].value_counts()
```

```python
#3 is not required now
```

```python
#4 divide into i/p and o/p 
#input is dimensional
x = df.iloc[:,2:4].values
x
```

```python
#output is 1 dimensional
y = df.iloc[:,4].values
y
```

```python
#5 train and test variable - train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state = 0)
#The variable names can change,  but the variable order can't change
#input_training, input_testing, output testing, output testing
```

```python
print(x.shape) #400rows and 2 cols - 100%
print(x_train.shape) #300rows and 2 cols - 75%
print(x_test.shape) #100 rows and 2 cols - 25%
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#6 normalize or scaling 
from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test) 
x_train
```

```python
x_train
```

```python
##7 Apply a classifier/regressor/cluster
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#8 Fitting the model
model.fit(x_train,y_train) #giving the data to the logistics Regression library
```

```python
#9 Predict the output
y_pred = model.predict(x_test)  #using the input testing values, we predict the output
y_pred #predicted values
```

```python
y_test #actual values
```

```python
#10 Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100 
```

```python
#LOGISTIC REGRESSION - MODEL 2 - IRIS FLOWER DATASET
#Datset - https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/IRIS.csv
```

```python
#1.Take the data and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/IRIS.csv')
df
```

```python
#2.Not required
#3.Not required
```

```python
#4.Divide the data into input and output
x = df.iloc[:,0:4].values
y = df.iloc[:,4].values
```

```python
#5.train and test varibales
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
#6.Not required because inputs are already scaled
```

```python
#7.apply a CLASSIFIER,regressor or clusterer
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#8.FItting the model
model.fit(x_train,y_train)
```

```python
#9.PREDICT THE OUTPUT
y_pred = model.predict(x_test)
y_pred #PREDICTED OUTPUT VALUES
```

```python
y_test # ACTUAL OUTPUT VALUES
```

```python
#10.Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
